```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load('prepared/all/windows/aaaaaarq_s017_t005_win0000.npy')

n = len(data)
cols = 6
rows = (n + cols - 1) // cols

plt.figure(figsize=(8, 4.5))  # smaller figure = tighter grid

for i in range(n):
    ax = plt.subplot(rows, cols, i + 1)
    ax.imshow(data[i], cmap='gray')
    ax.text(2, 10, f"Ch {i+1}", color='white', fontsize=7,
            bbox=dict(facecolor='black', alpha=0.4, pad=1))
    ax.set_xticks([])
    ax.set_yticks([])

# ZERO spacing between images
plt.subplots_adjust(
    left=0, right=1, top=1, bottom=0,
    wspace=0.01, hspace=0.01
)

plt.show()


```
